# Function: course_specific(course)
This function uses the data from getdata3 and chooses a random dish out of a meal list filtered based on the course chosen.

# Function: seasonal(month)
This function uses the data from getdata3() and chooses a random dish out of a meal list filtered based on the vegetables that are harvested at the given month, the current month by default. It uses the SeasonalVeggies() function to cycle the vegetables through the list of ingredients in every meal to determine which ones are seasonal. Based on said selection, it uses either downloadimage functions to get the photo and CreatePDF() function to create a PDF of the dish. If the function is successful, prints out 'Recipe Created!'

```python
WTC.seasonal()
```

# Function: combined(course, month)
This function uses the data from getdata3() and chooses a random dish out of a meal list filtered based on both vegetables harvested the given month as well as the course given. It uses SeasonalVeggies() to run the vegetables returned through the ingredients to determine the seasonality and then filters it further through said course. If there isn't a recipe that matches both criteria, it returns a string. If there are recipes present, it chooses a recipe randomly and produces a PDF of it.

```python
WTC.combined('dinner')
```
